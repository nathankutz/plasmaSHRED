### SVD for Plasmas

Low-rank features of data

```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from sklearn.utils.extmath import randomized_svd

plasma_data = sio.loadmat('data14fields/ne')
utemp = plasma_data['Data']
X = utemp - np.mean(utemp, axis=0)

n = (X.T).shape[0]
m = (X.T).shape[1]

print(X.shape)
print(n)
print(m)

u_ne,s_ne,v_ne=randomized_svd(X, n_components=20,n_iter='auto')

u_total = u_ne
v_total = v_ne
s_total = s_ne

k=20
usvd = u_ne[:, :k] @ np.diag(s_ne[:k]) @ v_ne[:k, :]

nx = 257
ny = 256

fig = plt.figure(figsize=(15, 10))
for k in range(6):
    Up = u_ne[:, k].reshape((nx, ny), order='F')
    uplot = Up.T
    ax = fig.add_subplot(3, 3, k+1)
    plt.savefig('u.png')
    axis('off')
    ax.imshow(uplot)
    
```

```python
fig = plt.figure(figsize=(15, 15))
for k in range(6):
    Up = u_ne[:, k].reshape((ny, nx), order='F')
    uplot = Up.T
    ax = fig.add_subplot(3, 3, k+1)
    ax.axis('off')
    ax.imshow(uplot,cmap='RdBu_r', interpolation='bilinear',)
    plt.savefig('u.pdf')
```

```python
fig = plt.figure(figsize=(15, 3))
ax = fig.add_subplot(2, 3, 1)

modesx = range(1,21,1)
plt.bar(modesx,100*s_ne/np.sum(s_ne), color='gray')


ax = fig.add_subplot(2, 3, 2); plt.plot(v_ne[0,:500], color='black')
ax = fig.add_subplot(2, 3, 3); plt.plot(v_ne[1,:500], color='black')
ax = fig.add_subplot(2, 3, 4); plt.plot(v_ne[2,:500], color='black')
ax = fig.add_subplot(2, 3, 5); plt.plot(v_ne[3,:500], color='black')
ax = fig.add_subplot(2, 3, 6); plt.plot(v_ne[4,:500], color='black')

plt.savefig('sig_v.pdf')
```
